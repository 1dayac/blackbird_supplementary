```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


df_simulated = pd.DataFrame(columns = ["Method", "SV type",  "Recall", "Precision", "F1"])
df_simulated = df_simulated.append({"Method" : "Blackbird", "SV type" : "Deletions", "Precision" : 0.940, "Recall" : 0.738, "F1" : 0.827}, ignore_index = True)
df_simulated = df_simulated.append({"Method" : "Blackbird", "SV type" : "Insertions", "Precision" : 0.960, "Recall" : 0.586, "F1" : 0.728}, ignore_index = True)

df_insertions = df_simulated[df_simulated["SV type"] == "Insertions"]
df_deletions = df_simulated[df_simulated["SV type"] == "Deletions"]
```

```
fig, axs = plt.subplots(1, 2, figsize =(20,10))
font = {'family' : 'normal',
        'size'   : 14}
plt.rc('font', **font)
plt.rcParams['figure.dpi'] = 400
df_deletions.set_index("Method").plot(kind = "bar", ax = axs[0], ylim=(0,1.0), yticks = np.arange(0, 1.01, 0.1), title = "Deletions", xlabel = "", rot = 0)
df_insertions.set_index("Method").plot(kind = "bar", ax= axs[1], ylim=(0, 1.0), yticks = np.arange(0, 1.01, 0.1), title = "Insertions", xlabel = "", rot = 0)
axs[0].get_legend().remove()
fig.show()
```
